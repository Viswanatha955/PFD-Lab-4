```python
Mean = np.mean(S)
Median = np.median(S)
S1 = pd.Series(data=S)
Mode = S1.mode()
print('Mean:', Mean)
print('Median:', Median)
print('Mode:', Mode[0])
```

```python
# Quartiles
Q1 = np.percentile(S, 25)
Q2 = np.median(S)
Q3 = np.percentile(S, 75)

# Interquartile Range (IQR)
IQR = Q3 - Q1
```
